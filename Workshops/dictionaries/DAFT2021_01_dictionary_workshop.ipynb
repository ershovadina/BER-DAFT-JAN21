# discuss when you would use a dict and when a list

* list:
    * when we need to store values that we want to iterate over
    * that we want to address (with indexing e.g.)
    * we want to have access to information about that list (length, maximum, minimum, average, etc.)
    * the elements are not in need of a key / description. That means, we don't primarily care for retrieving the elements out of this list (althoug it is possible) 
    
* dictionary
    * when we need to store something that we want to iterate over
    * that we want to address but with exact "keys"
    * when we need to have **structured data** (key-value-pairs)
        * e.g. you want to store the score of your 5 machine learning models that you have trained. key: name of the model, value: the score itself
    * structured data, that we need to retrieve at some point, by providing the key and getting the value
    * with regards what we do now: JSON, relational databases (because they need nested dictionaries)
